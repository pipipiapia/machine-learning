```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as la
```

```python
def get_m1m2(X, y):
    """
    X is data matrix samples in rows, y is class, -1 or 1
    line joining means
    """
    #get means and covariances of both classes
    X_1 = X[np.where(y>0)[0]]
    X_2 = X[np.where(y<=0)[0]]
    m1 = np.mean(X_1, axis=0)
    m2 = np.mean(X_2, axis=0)
    return (m1 - m2)

def plot_hist(X, y, w, data, label):
    """takes in projected data dn plots histogram - just needs c"""
    c = np.inner(X, w)/np.inner(w,w)
    plt.figure()
    plt.hist(c[np.where(y>0)[0]], color='blue', alpha=0.3)
    plt.hist(c[np.where(y<=0)[0]], color='green', alpha=0.3)
    plt.savefig(f"./images/fld_hist_{data}_{label}.png", bbox_inches="tight")
    plt.show()
    plt.close()
#     print(c)
    
def plot_ld(X, y, w, data, label):
    X_1 = X[np.where(y>0)[0]]
    X_2 = X[np.where(y<=0)[0]]
#     print(X_1.shape, X_2.shape)
    proj = np.outer(w,w)/np.inner(w,w)
    X_1p = np.dot(X_1, proj)
    X_2p = np.dot(X_2, proj)
#     plt.plot(X_1p[:,1], X_1p[:,2], 'b') 
#     plt.plot(X_2p[:,1], X_2p[:,2], 'g.')
    plt.axline((0, w[0]), slope=w[2]/w[1], color='r', linestyle=(0, (5, 5)))
    plt.plot(X_1[:,1], X_1[:,2], 'b.')
    plt.plot(X_2[:,1], X_2[:,2], 'g.')
    plt.xlabel(r"$x_1$")
    plt.ylabel(r"$x_2$")
    plt.savefig(f"./images/fld_{data}_{label}.png", bbox_inches="tight")
def get_fld(X, y):
    """
    X is data matrix samples in rows, y is class, -1 or 1
    outputs best discriminant vector w
    """
    #get means and covariances of both classes
    X_1 = X[np.where(y>0)[0]]
    X_2 = X[np.where(y<=0)[0]]
    m1 = np.mean(X_1, axis=0)
    m2 = np.mean(X_2, axis=0)
    C1 = np.cov(X_1.T)
    C2 = np.cov(X_2.T)
    Cw = C1 + C2

    return np.dot(la.pinv(Cw), (m1 - m2))
```

```python
N = 100
X3 = np.vstack([np.random.randn(N//2, 2)/2-.5,np.random.randn(N//2, 2)/2+.5 ])
y3 = np.hstack([np.ones(N//2)*-1, np.ones(N//2)]).reshape(-1, 1)
X3 = np.concatenate([np.ones(N).reshape(-1,1), X3], axis=1)

n = 1000
C = [[1,0.98],[0.98,1]]
A = la.cholesky(C)
Xa = np.random.randn(n//2,2)
Xa = np.dot(Xa, A)
Xb = np.random.randn(n//2,2)
Xb = np.dot(Xb, A) +1


print(Xa.shape, Xb.T.shape)
X4 = np.vstack([Xa, Xb])
X4 = np.concatenate([np.ones(n).reshape(-1,1), X4], axis=1)
# print(X4.shape)
y4 = np.hstack([np.ones(n//2), np.zeros(n//2)])
print(X4.shape, y4.shape)
```

```python
w3 = get_fld(X3, y3)
print(w3)
plot_ld(X3, y3, w3, 'X3', 'fld')
plot_hist(X3, y3, w3, 'X3', 'fld')
```

```python
w4 = get_fld(X4, y4)
plot_ld(X4, y4, w4, 'X4', 'fld')
plot_hist(X4, y4, w4, 'X4', 'fld')
```
